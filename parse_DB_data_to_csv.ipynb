```python
access_token ='EAACEdEose0cBAOxkm9grcaCa5P5DW6XrwW59ktudZA1hfzxttS0jWUDkoZAWgG7mzZCd6vFw3O10hibrNLlZCTVvIBcTBKQy9D4GizMH23dSl8RzBTq0dJDAJ2mSSIsAuZABZCqUe4FbuNn5k8FjuscXIlbyfTilZBaHppdNuxurCylptm85Hf4'
group_id = '400948843346851'

# get the data first time
response = requests.get('https://graph.facebook.com/v2.3/{}/feed?access_token={}'.format(group_id, access_token))

res_data = response.json()
results = []

time = 0
while res_data['paging']['next'] != None:
    if time < 50:
        data = res_data['data']
        result = []

        for n, d in enumerate(data):
            message = dict()
            try:
                message['message'] = d['message']
            except:
                # no messages
                pass

            try:
                message['link'] = d['link']
            except:
                # no shared link
                pass

            message['from'] = d['from']['name']
            message['time'] = d['created_time']

            try:
                comments = []
                for com in d['comments']['data']:
                    comment = dict()
                    comment['message'] = com['message']
                    comment['from'] = com['from']['name']
                    comment['time'] = com['created_time']
                    comments.append(comment)

                message['comments'] = comments
            except:
                # no comments
                pass
            result.append(message)


        results += result

        # get data from next page
        time += 1
        print('fetch {} page'.format(time))
        response = requests.get(res_data['paging']['next'])
        res_data = response.json()
    else:
        break


doc = Document()

for r in results:
    try:
        doc.add_heading(r['message'], level=2)
    except:
        pass
    
    try:
        doc.add_heading(r['link'], level=2)
    except:
        pass
    
    doc.add_heading(r['from'] + ' ' + r['time'], level=3)
    
    try:
        for x in r['comments']:
            name = doc.add_paragraph(x['from'])
            name.style.font.bold = True
            name.add_run('   ')
            name.add_run(x['message']).bold = False
    except:
        pass
    doc.add_paragraph('--------')
    
doc.save('ç³–å°¿ç—…äº’å‹•ç¶²ç¤¾åœ˜è¨Šæ¯.docx')
```
