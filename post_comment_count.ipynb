```python
import json
```

```python
file = open('./diabetes-interaction-net-data.json')
```

```python
raw_data = file.readline()
```

```python
json_data = json.load(file)
```

```python
days_data = json_data['data']
```

```python
total_post = []
for day_data in days_data:
    for post in day_data:
        total_post.append(post)
```

```python
len(days_data[0][0])
```

```python
users = {}
for post in total_post:
    poster = post['from']
    print(poster)
    if poster['id'] not in users.keys():
        users[poster['id']] = {
            'name': poster['name'],
            'post_count': 1,
            'comment_count': 0
        }
    else:
        user_data = users[poster['id']]
        user_data['post_count'] += 1
        
    try:
        comments = post['comments']['data']
        for comment in comments:
            commenter = comment['from']
            if commenter['id'] not in users.keys():
                users[commenter['id']] = {
                    'name': commenter['name'],
                    'post_count': 0,
                    'comment_count': 1
                }
            else:
                user_data = users[commenter['id']]
                user_data['comment_count'] += 1
    except KeyError:
        print('no comment')
```
